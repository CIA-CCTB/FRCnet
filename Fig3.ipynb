## Monitoring of training and validation loss for different configurations

----

Load dependencies:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
%matplotlib inline
```

Open training logs for different preprocessing:

```python
data1 = pd.read_csv('data/trace_frc_ssim_nokde.csv')

x1 = data1['step']
x1 = x1/1000

y1 = data1['gen_loss_frc_int']
```

```python
data2 = pd.read_csv('data/trace_frc_ssim.csv')

x2 = data2['step']
x2 = x2/1000

y2 = data2['gen_loss_frc_int']
```

Plot FRC loss trained with isotropic Gaussian vs. anisotropic KDE:

```python
plt.figure(figsize=(3.5,3))
plt.plot(x1,y1, label = 'isotropic KDE', color="silver")
plt.plot(x2,y2, label = 'anisotropic KDE', color="black")
plt.xlabel('Epoch')
plt.ylabel('FRC Loss')
plt.legend()
plt.show()
```

Open training logs for different loss functions:

```python
path1 = ''
path2 = 'data/trace_ssim_only.csv'
path3 = ''

data1 = pd.read_csv('data/trace_frc_only.csv')
x1 = data1['step']
x1 = x1/1000
y1s = data1['gen_loss_SSIM']
y1f = data1['gen_loss_frc_int']

data2 = pd.read_csv('data/trace_ssim_only.csv')
x2 = data2['step']
x2 = x1/1000
y2s = data2['gen_loss_SSIM']
y2f = data2['gen_loss_frc_int']

data3 = pd.read_csv('data/trace_frc_ssim.csv')
x3 = data1['step']
x3 = x1/1000
y3s = data3['gen_loss_SSIM']
y3f = data3['gen_loss_frc_int']
```

Plot MS-SSIM loss for different training loss configurations:

```python
plt.figure(figsize=(3.5,3))
plt.plot(x1[5:],y1s[5:], label = 'FRC only', color="red")
plt.plot(x1[5:],y2s[5:], label = 'SSIM only')
plt.plot(x1[5:],y3s[5:], label = 'FRC + SSIM', color="black")
plt.xlabel('Epoch')
plt.ylabel('MS-SSIM')
plt.legend()
plt.title("Figure 3e")
plt.show()
```

Plot FRC loss for different training loss configurations:

```python
plt.figure(figsize=(3.5,3))
plt.plot(x1[5:],y1f[5:], label = 'FRC only', color="red")
plt.plot(x1[5:],y2f[5:], label = 'SSIM only')
plt.plot(x1[5:],y3f[5:], label = 'FRC + SSIM', color="black")
plt.xlabel('Epoch')
plt.ylabel('FRC Loss')
plt.legend()
plt.title("Figure 3f")
plt.show()
```
